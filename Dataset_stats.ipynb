```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Content of DB
labels = ['Political', 'Unreliable', 'Bias', 'Fake', 'Conspiracy','Unknown','Rumor','Clickbait','Junksci','Satire','Reliable','Hate']
Authors_com = [1795907, 424652, 444875, 210863, 321330,232870,73732,48817,63526,44329,19286,9325]
Authors_com = [round(item/sum(Authors_com)*100) for item in Authors_com]
Authors_dif = [16993, 781, 7103, 3622, 5530,8803,13079,2162,2130,542,675,1263]
Authors_dif = [round(item/sum(Authors_dif)*100) for item in Authors_dif]
Domains = np.array([51,25,71,21,62,82,1,14,19,34,3,11])
Articles = np.array([289666, 141614,136381,123929,109793,47413,46524,21723,17408,14263,6601,3619])
Authors_dif2 = np.array([16993, 781, 7103, 3622, 5030,8803,13079,2162,2130,542,675,1253])

x = np.arange(len(labels))  # the label locations
width = 0.45  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, Authors_com, width, label='Total')
rects2 = ax.bar(x, Authors_dif, width, label='Distinct')
#rects3 = ax.bar(x + width, Authors_dif, width, label='Domains')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('% of total')
ax.set_title('Distribution of authors total vs distinct')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylim(0,54)
plt.xticks(rotation=-45)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
